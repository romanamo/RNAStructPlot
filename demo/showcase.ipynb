# Showcase

A quick introduction to plot types of RNAStructPlot.

## Parse Structure

The structure can be created using its rna-sequence and its dotbracket-notation.

```julia
using RNAStructPlot
using CairoMakie

rnastruct = Parse.dotbracketbase(
    "CCGAUAGAGGCGUGCGGUCAAGGUCCGG", 
    "(((((...(((.....)))...)).)))"
)
```

## Circle Plot

with bonds as straight lines

```julia
f = Figure()
ax = Axis(f[1,1], title="Circle Plot (straight lines)", aspect=DataAspect())

Recipes.circleplot!(f[1,1], rnastruct)

hidedecorations!(ax)
hidespines!(ax)

f
```

with bonds as curves

```julia
f = Figure()
ax = Axis(f[1,1], title="Circle Plot (curved lines)", aspect=DataAspect())

Recipes.circleplot!(f[1,1], rnastruct, layout=:curves)

hidedecorations!(ax)
hidespines!(ax)

f
```

## Arc Plot


```julia
f = Figure()
ax = Axis(f[1,1], title="Arc Plot", aspect=DataAspect())

Recipes.arcplot!(f[1,1], rnastruct)

hidedecorations!(ax)
hidespines!(ax)

f
```

## Structure Plots



The RNA can be drawn with loops as regular polygons and rectangles as stem regions:

```julia
f = Figure()
ax = Axis(f[1,1], title="Polygonal Struct Plot", aspect=DataAspect())

Recipes.structureplot!(f[1,1], rnastruct, layout=:polygonal)

hidedecorations!(ax)
hidespines!(ax)

f
```

Using the polygon methode can have its problems. Sometimes the resulting drawing contains crossing/overlaps. The modified version provides a fallback, which guarantees a drawing without crossing/overlaps.

```julia
f = Figure()
ax = Axis(f[1,1], title="Modified Struct Plot", aspect=DataAspect())

Recipes.structureplot!(f[1,1], rnastruct, layout=:modified)

hidedecorations!(ax)
hidespines!(ax)

f
```

```julia
sequence = "CGCUUCAUAUAAUCCUAAUGAUAUGGUUUGGGAGUUUCUACCAAGAGCCUUAAACUCUUGAUUAUGAAGUG"
notation = "...(((((((..((((((.........))))))......).((((((.......))))))..))))))..."
rnastruct2 = Parse.dotbracketbase(sequence, notation)
```

```julia
f = Figure()
ax = Axis(f[1,1], title="Modified Struct Plot", aspect=DataAspect())

Recipes.structureplot!(f[1,1], rnastruct2, layout=:modified)

hidedecorations!(ax)
hidespines!(ax)

f
```
